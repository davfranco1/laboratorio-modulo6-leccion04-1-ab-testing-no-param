 Pasos a seguir:

```python
lista_grupos = list(df["Education"].unique())
lista_grupos
```

```python
valores = sp.crear_df_grupos(df, "Education", "Flights Booked")
sp.elegir_test(valores)
```

```python
sns.barplot(df, x="Education", y="Flights Booked")
```

## Ejercicio 3:
Determina si existe una diferencia significativa en el salario de los clientes en función del género de los mismos.

Pasos a seguir:
   - Definición de la hipótesis.
   - Prueba de hipótesis
   - Análisis de resultados
   - Conclusiones

Planteamiento de hipótesis:
- H0: no existe diferencia significativa en el salario en función del género.
- H1: existe diferencia significativa en el salario en función del género.

```python
df_filtrado_salarios.isna().sum()
```

```python
df_filtrado_salarios.shape
```

```python
sp.iteraciones_normalidad(df_filtrado_salarios,"Gender","Salary")
```

Dado que no tenemos una distribución normal, se aplicará un test paramétrico. Con grupos independientes, haremos uso de Mann-Whitney.

```python
valores = sp.crear_df_grupos(df_filtrado_salarios, "Gender", "Salary")
sp.elegir_test(valores, dependencia=False)
```

```python
sns.boxplot(df_filtrado_salarios, x="Gender", y="Salary")
```

```python
sp.hisplot_grupos(df_filtrado_salarios, "Gender", "Salary")
```

```python
sns.barplot(df_filtrado_salarios, x="Gender", y="Salary")
plt.ylim(79200, 79800)
```

```python
df_filtrado_salarios.groupby("Gender")["Salary"].mean()
```

- Haciendo uso del test Mann-Whitney obtenemos un p-value menor a 0.05, contndo con evidencia estadística suficiente para rechazar la hipótesis nula, lo que implica que hay diferencias significativas entre los salarios por género.

- A pesar de que, en el último cálculo podemos observar que, las medias son muy similares, y las distribuciones siguen ese mismo patrón, esa diferencia se debería a que tenemos colas muy grande al extremo de la gráfica.

- Para ser capaces de observar mejor las diferencias arrojadas por el estadístico, sería necesario coger una muestra de la parte en que se concentran la mayoría de los valores, obteniendo una distribución más normal, a la que volveríamos a aplicar los cálculos.
